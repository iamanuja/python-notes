```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
y = np.array([1,2,3,4])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = [ "hotpink","Yellow", "red", "green"]
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True,colors=mycolors,autopct = '%.2f')
plt.legend(loc ="lower left") # loc ="lower right <= by default
plt.show() 
```

```python
y = np.array([1,2,3,4])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = [ "hotpink","Yellow", "red", "green"]
plt.pie(y, labels = mylabels, explode = myexplode,startangle=90, shadow = True,colors=mycolors,autopct = '%.2f')
plt.legend(loc ="lower left") # loc ="lower right <= by default
plt.show() 
```

## box and whisker plot. 

```python
dt=sns.load_dataset("tips")
dt.shape
```

```python
dt.head(5)
```

```python
sns.boxplot(dt, x="total_bill" ,y="time")#,hue="sex")
```

# Displot

```python
sns.displot(dt,x="total_bill",col="time",kde=True)
```

```python
dt["total_bill"].describe()
```

```python
sns.displot(dt,x="total_bill") #'hist', 'kde', 'ecdf' # kde=True
```

```python
sns.displot(dt, x="total_bill", kind="kde",hue="time",col="sex",height=4, aspect=.8,)
```
